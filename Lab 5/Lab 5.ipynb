```python
import pandas as pd

df=pd.read_csv('Chromo.txt', sep='\t')
df_new_1=df[(df['Time (min)']>=4)  ]
df_new=df_new_1[(df['Time (min)']<=7)]
df_new.plot('Time (min)','Value (mAU)', ylabel='Value (mAU)', title='Chromo(4-7 min)')
```
