```python
#упражнение  3
import os
import matplotlib.pyplot as plt
import pandas as pd

def reading_x (name):
    f=open(name, 'r')
    Flag=True
    x=[]
    for line in f: 
        if Flag==False:
            a=line.split(',')
            x.append(float(a[0]))
        Flag=False
    f.close()  
    return x

def reading_y (mass, name):
    f=open(name, 'r')
    Flag=True
    new_mass=mass
    for line in f: 
        if Flag==False:
            a=line.split(',')
            for i in range (len(mass)):
                new_mass[i].append(float(a[i]))
        Flag=False
    f.close()  
    return new_mass   
    
way=r'C:\Users\mefim\infa_2021_efimova\Lab8'
files = os.listdir(way)

for file in files:
    way, file_type = os.path.splitext(file)
    if file_type == '.csv':
        df = pd.read_csv(file)
        column = df.shape[1]-1
        mass = [[] for i in range(column)]
        x=reading_x(file)
        new_mass = reading_y(mass, file)
        for i in range (len(new_mass)):
            plt.scatter(x, new_mass[i])
            plt.xlabel('x')
            plt.ylabel('y')
            plt.title(file)
        plt.savefig(file+'.png')
        plt.show()
       
```
