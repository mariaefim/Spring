```python
#упражнение 2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

students = pd.read_csv("StudentsPerformance.csv")
type = ('reading score', 'math score', 'writing score')
plt.figure(figsize=(40,20))
colors =['green', 'orange', 'red']
number = students.shape[0]
x=[]
for i in range(number):
    x.append(i+1)
            
for i in range(len(type)):
    plt.scatter(x, students[type[i]], c=colors[i])
    plt.title('Results', fontsize = 45)
    plt.xlabel('students number', fontsize = 35)
    plt.ylabel('score', fontsize = 35)
    plt.legend(type, fontsize = 35)
    
    
plt.show()
```

```python
# упражнение 3
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('Chromo.txt', sep='\t')
df_new_1=df[(df['Time (min)']>=4)  ]
df_new=df_new_1[(df['Time (min)']<=7)]
df_new.plot('Time (min)','Value (mAU)', ylabel='Value (mAU)', title='Chromo(4-7 min)')

maximum =max(df[df.columns[1]])
ind = df[df[df.columns[1]] == maximum].index.values.astype(int) #индекс max значения 
xmean=df[df.columns[0]][int(ind)]

plt.arrow(x=xmean, y=25, dx=0, dy=-(25-maximum-1.2),width= 0.06,  head_length=1, head_width=0.2)
plt.text(x=5.4,y= 25.3,s='max')
```
